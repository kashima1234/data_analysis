```python
for dtype in['float','int','object']:
    selected_dtype=ds.select_dtypes(include=[dtype])
    mean_usage_b=selected_dtype.memory_usage(deep=True).mean()
    mean_usage_mb=mean_usage_b/1024
    print("Average memory usage for {} columns: {:03.3f} КВ".format(dtype,mean_usage_mb))
```

```python
def mem_usage(pandas_obj):
    if isinstance(pandas_obj,pd.DataFrame):
        usage_b=pandas_obj.memory_usage(deep=True).sum()
    else:
        usage_b=pandas_obj.memory_usage(deep=True)
    usage_mb=usage_b/1024
    return "{:03.3f} КВ".format(usage_mb)
```

```python
ds_int=ds.select_dtypes(include=['int'])
converted_int = ds_int.apply(pd.to_numeric,downcast='unsigned')

print(mem_usage(ds_int))
print(mem_usage(converted_int))

compare_ints=pd.concat([ds_int.dtypes,converted_int.dtypes],axis=1)
compare_ints.columns=['before','after']
compare_ints.apply(pd.Series.value_counts)
```

```python
ds_float=ds.select_dtypes(include=['float'])
converted_float = ds_float.apply(pd.to_numeric,downcast='float')

print(mem_usage(ds_float))
print(mem_usage(converted_float))

compare_floats=pd.concat([ds_float.dtypes,converted_float.dtypes],axis=1)
compare_floats.columns=['before','after']
compare_floats.apply(pd.Series.value_counts)
```

```python
optimized_ds=ds.copy()

optimized_ds[converted_int.columns]=converted_int
optimized_ds[converted_float.columns]=converted_float

print(mem_usage(ds))
print(mem_usage(optimized_ds))
```

```python
ds_obj=ds.select_dtypes(include=['object']).copy()
ds_obj.describe().T
```

```python
# Convert categorical columns to 'category' dtype if the unique value ratio is low
converted_obj = pd.DataFrame()
for col in ds_obj.columns:
    num_unique_values = len(ds_obj[col].unique())
    num_total_values = len(ds_obj[col])
    
    if num_unique_values / num_total_values < 0.5:
        converted_obj.loc[:, col] = ds_obj[col].astype('category')
    else:
        converted_obj.loc[:, col] = ds_obj[col]
```

```python
print(mem_usage(ds_obj))
print(mem_usage(converted_obj))

# Convert categorical columns back to string to avoid sorting issues in value_counts
converted_obj = converted_obj.apply(lambda x: x.astype(str) if isinstance(x.dtype, pd.CategoricalDtype) else x)

# Compare dtypes before and after conversion
compare_obj = pd.concat([ds_obj.dtypes, converted_obj.dtypes], axis=1)
compare_obj.columns = ['before', 'after']

# Apply value_counts without errors
print(compare_obj.apply(pd.Series.value_counts))
```

```python
optimized_ds[converted_obj.columns]=converted_obj
mem_usage(optimized_ds)
```
